# Random walks and Brownian motion

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_walk_1d(x):
    plt.plot(x)
    plt.xlabel("step number")
    plt.ylabel("value")
```

```python
def gen_random_walk(x0, n_step=100):
        """
        """     
        w = np.ones(n_step) * x0
        
        for i in range(1, n_step):
            # Sampling from the 
            yi = np.random.choice([1,-1])
            # Weiner process
            w[i] = w[i-1]+(yi/np.sqrt(n_step))
        
        return w
```

```python
def gen_normal(x0, n_step=100):
        """
        """

        w = np.ones(n_step) * x0
        
        for i in range(1,n_step):
            # Sampling from the Normal distribution
            yi = np.random.normal()
            # Weiner process
            w[i] = w[i-1]+(yi/np.sqrt(n_step))
        
        return w
```

```python
x0=0
w = gen_random_walk(x0)

print(w)
plot_walk_1d(w)

w = gen_normal(x0)
print(w)
plot_walk_1d(w)
```
